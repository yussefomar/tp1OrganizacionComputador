```python
#¿cual es la zona mas costosa en promedio?
prop_vent.groupby('state_name').agg({'price_aprox_usd':'mean'}).sort_values( by='price_aprox_usd' ,ascending=False).plot(kind='bar')
```

```python
#¿barrios mas costosos Capital Federal?
capital = prop_vent.dropna(subset=['price_aprox_usd'])
top_mayor_prom_pre_cap = capital[['state_name','place_name','price_aprox_usd']][capital['state_name']=="Capital Federal"]\
.groupby('place_name').agg({'price_aprox_usd':'mean'})\
.sort_values(by='price_aprox_usd',ascending=False)
top_mayor_prom_pre_cap[:10].plot(kind='bar')
```

```python
#hay alguna relacion entre el precio y la cantidad de departamentos por area cubierta en capital federal
x=capital['state_name']=="Capital Federal"
y=capital['property_type']=="apartment"
z=capital['surface_covered_in_m2']<80.000000
relacion_cantidad_departamentos_con_precio_zona_en_capital=top_con_mayor_promedio_precio_GB=\
capital[['state_name','place_name','price_aprox_usd','property_type']][ x& y&z].groupby('place_name').\
agg({'price_aprox_usd':'mean','property_type':'count'}).sort_values( by='price_aprox_usd' ,ascending=False)

relacion_cantidad_departamentos_con_precio_zona_en_capital.rename(columns={'property_type': 'Cantidad De Departamentos'}, inplace=True)
```

```python
relacion_cantidad_departamentos_con_precio_zona_en_capital.plot.scatter(x='Cantidad De Departamentos',y='price_aprox_usd',label='Relacion del precio promedio por estado cantidad de departamentos')
```

```python
#Evolucion del precio de la propiedad a travez zel ano 2012 a 2017 en una determinada zona
prop_fecha=prop_vent['created_on'].str.split(' ',3,expand=True) 
a=prop_fecha[0].str.split('-',3,expand=True)
 
prop_vent['year']=a[0]
prop_vent['month']=a[1]
prop_vent['day']=a[2]
prop_vent =prop_vent
```

```python
#cantidad de publicaciones de ventas,durante el 2012 a 2017
%matplotlib inline

fig = plt.figure(figsize=(10,4));

ax = fig.add_axes([0,0,1,1]);
prop_vent.groupby('year').count()['created_on'].plot.line(c='b',label="buenos aires y alrededores")

 #cantidad de publicaciones de ventas,durante el 2012 a 2017 solo en Capital federal
 
x=prop_vent['state_name']=="Capital Federal"
prop_vent[x].groupby('year').count()['created_on'].plot.line(c='g', label="Capital Federal") 

 
ax.legend();
```

```python
#cantidad de superficie cubierta de la empresa properati en capital federal,Bs.As. G.B.A. Zona Norte,Bs.As. G.B.A. Zona Oeste
 
prop_vent[['surface_covered_in_m2','state_name']].groupby('state_name').agg({'surface_covered_in_m2':'sum'}).plot(kind='bar')
```

```python
#cantidad de publicaciones de ventas,durante el 2012 a 2017 solo en "Bs.As. G.B.A. Zona Sur"
gbazs=prop_vent['state_name']=="Bs.As. G.B.A. Zona Sur"
 
c=prop_vent[['year','state_name']][gbazs].groupby('year').agg({'year':'count'})
c.plot()
```

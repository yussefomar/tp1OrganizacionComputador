```python
%matplotlib inline

import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('default') # haciendo los graficos un poco mas bonitos xD
plt.rcParams['figure.figsize'] = (15, 5)
file_directory = './data_file/properati-AR-2017-08-01-properties-sell.csv'
prop = pd.read_csv(file_directory, low_memory=False)
#prop.groupby('price')
prop.tail()
```

# se eliminan las columnas que se consideran innecesarias


```python
prop.info()
```

```python
#se nos pode quedarnos con las propiedades de Argentina de cap. fed. y gba
col_relevantes = ["created_on", "property_type", "place_name", "state_name", "lat", "lon", "price", "currency",\
                 "price_aprox_local_currency", "price_aprox_usd", "surface_total_in_m2", "surface_covered_in_m2",\
                 "floor", "rooms", "expenses", "description", "title"]
prop_baires = prop.loc[(prop["country_name"] == "Argentina") & ( (prop['state_name'] == "Capital Federal") |\
                                                                 (prop['state_name'].str.contains("G.B.A")) |\
                                                                (prop['state_name'].str.contains("Buenos Aires")) )\
                             ,col_relevantes]
prop_baires['state_name'].value_counts()
```

```python
# definimos las columnas usadas para definir precio promedio por zona
col_zonas_precio = ['place_name', 'price']

# Vemos precio promedio de Cap. Fed.
top_mayor_prom_pre_cap = prop_baires.loc[prop_baires['state_name']=="Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
top_mayor_prom_pre_cap
```

```python
# Vemos precio promedio de buenos aires (provincia)
top_mayor_prom_pre_bsas = prop_baires.loc[prop_baires['state_name'] != "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
top_mayor_prom_pre_bsas
```

```python
# apartamentos más caros de cap. fed.
pro_deptos = prop_baires.loc[prop_baires['property_type'] == 'apartment', :]
deptos_mayor_prom_pre_capfed = pro_deptos.loc[pro_deptos['state_name'] == "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
deptos_mayor_prom_pre_capfed
```

```python
# apartamentos más caros de buenos aires
deptos_mayor_prom_pre_bsas = pro_deptos.loc[pro_deptos['state_name'] != "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
deptos_mayor_prom_pre_bsas
```

```python
# locales más caros de cap. fed.
pro_stores = prop_baires.loc[prop_baires['property_type'] == 'store', :]
stores_mayor_prom_pre_capfed = pro_stores.loc[pro_stores['state_name'] == "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
stores_mayor_prom_pre_capfed
```

```python
# locales más caros de buenos aires
stores_mayor_prom_pre_bsas = pro_stores.loc[pro_stores['state_name'] != "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
stores_mayor_prom_pre_bsas
```

```python
# PH más caros de cap. fed.
pro_ph = prop_baires.loc[prop_baires['property_type'] == 'PH', :]
ph_mayor_prom_pre_capfed = pro_ph.loc[pro_ph['state_name'] == "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
ph_mayor_prom_pre_capfed
```

```python
# PH más caros de buenos aires
ph_mayor_prom_pre_bsas = pro_ph.loc[pro_ph['state_name'] != "Capital Federal", :]\
[col_zonas_precio].groupby('place_name').agg({'price':'mean'}).sort_values(by='price',ascending=False)
ph_mayor_prom_pre_bsas
```

```python
#los lugares con los ph mas costosos asi tmb se puede hacer con las casas mas costosas o los locales mas costosos
esPH=prop_baires['property_type'] == "PH"
esBS= prop_baires['state_name'] != "Capital Federal"
ph = prop_baires.loc[esPH & esBS]
ph.groupby('place_name').agg({'price_aprox_usd':'mean'})\
.sort_values(by='price_aprox_usd',ascending=False)
```
